```python
# Install required packages (run this cell first in Colab)
!pip install nltk scikit-learn pandas numpy matplotlib seaborn

import pandas as pd
import numpy as np
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import SnowballStemmer
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import warnings
warnings.filterwarnings('ignore')

# Download required NLTK data
nltk.download('punkt')
nltk.download('stopwords')

class DomesticViolenceClassifier:
    def __init__(self):
        self.stemmer = SnowballStemmer('spanish')
        self.stop_words = set(stopwords.words('spanish'))

        # Enhanced keyword dictionaries with negation patterns
        self.high_risk_keywords = {
            'weapons': ['arma', 'disparo', 'revolver', 'tiros', 'pistola', 'puñal', 'bala',
                       'cuchillo', 'machete', 'navaja', 'escopeta', 'rifle', 'explosivo',
                       'cortopunzante', 'blanca', 'fuego'],
            'violence': ['golpe', 'golpeando', 'golpear', 'agresion', 'agredir', 'atacar',
                        'matar', 'asesinar', 'violencia', 'brutal', 'salvaje', 'torturar',
                        'patada', 'puñetazo', 'bofetada', 'estrangular', 'ahorcar'],
            'threats': ['amenaza', 'amenazar', 'intimidar', 'extorsion', 'chantaje',
                       'muerte', 'matar', 'destruir', 'quemar'],
            'injuries': ['herida', 'herido', 'lesion', 'sangre', 'sangrado', 'ambulancia',
                        'fractura', 'moretones', 'contusion', 'traumatismo', 'hospitalizar',
                        'medico', 'primeros', 'auxilios', 'trauma', 'lesiones'],
            'vulnerable': ['embarazada', 'gestante', 'niñas', 'niños', 'menor', 'anciano',
                          'discapacitado', 'bebe', 'infante', 'adolescente', 'gestación'],
            'substances': ['embriaguez', 'alicorado', 'droga', 'narcótico', 'sustancia',
                          'alcohol', 'borracho', 'intoxicado', 'ebrio', 'licor'],
            'legal': ['medida', 'proteccion', 'denuncia', 'orden', 'restriccion', 'judicial',
                     'violacion', 'incumplimiento', 'antecedentes'],
            'severity': ['urgente', 'grave', 'critico', 'emergencia', 'inmediato', 'extremo',
                        'peligro', 'riesgo', 'alto', 'severo']
        }

        self.moderate_risk_keywords = {
            'verbal': ['grito', 'gritando', 'insulto', 'insultando', 'discusion', 'pelea',
                      'verbal', 'altercado', 'disputa', 'discutir'],
            'psychological': ['psicologico', 'mental', 'emocional', 'humillacion', 'control',
                             'aislamiento', 'manipulacion', 'intimidacion'],
            'property': ['daño', 'romper', 'quebrar', 'destruir', 'vandalismo', 'propiedad',
                        'objetos', 'casa', 'muebles'],
            'presence': ['presencia', 'acechar', 'seguir', 'hostigar', 'molestar', 'perseguir']
        }

        self.low_risk_keywords = {
            'resolved': ['calmo', 'tranquilo', 'resuelto', 'acuerdo', 'reconciliacion',
                        'solucionado', 'controlado'],
            'minor': ['menor', 'leve', 'pequeño', 'insignificante'],
            'preventive': ['preventivo', 'precaucion', 'informativo', 'orientacion']
        }

        # Negation patterns in Spanish
        self.negation_patterns = [
            r'\b(no|sin|nunca|nada|nadie|ninguno|ninguna|ni)\s+',
            r'\b(n|neg|negativo)\s+',
            r'\b(ausencia|falta)\s+de\s+',
            r'\b(no\s+se|no\s+hay|no\s+hubo|no\s+tiene|no\s+presenta)\s+'
        ]

        # Abbreviation patterns common in police reports
        self.abbreviations = {
            'PONAL': 'policia nacional',
            'dv': 'violencia domestica',
            'vic': 'victima',
            'agr': 'agresor',
            'loc': 'localidad',
            'dir': 'direccion',
            'tel': 'telefono',
            'obs': 'observaciones',
            'ref': 'referencia',
            'aprox': 'aproximadamente',
            'h': 'heridos',
            'm': 'muertos',
            'fem': 'femenino',
            'masc': 'masculino',
            'pers': 'persona',
            'suj': 'sujeto'
        }

        self.vectorizer = None
        self.model = None

    def preprocess_text(self, text):
        """Enhanced text preprocessing with negation handling"""
        if pd.isna(text):
            return ""

        text = str(text).lower()

        # Expand abbreviations
        for abbr, full in self.abbreviations.items():
            text = re.sub(r'\b' + abbr + r'\b', full, text)

        # Handle common transcription errors
        text = re.sub(r'\bn\b', 'no', text)  # Common abbreviation for "no"
        text = re.sub(r'\bs/\b', 'sin', text)  # Common abbreviation for "sin"

        # Remove punctuation but keep spaces
        text = re.sub(r'[^\w\s]', ' ', text)

        # Remove extra whitespace
        text = re.sub(r'\s+', ' ', text).strip()

        return text

    def detect_negation_context(self, text):
        """Detect negated phrases and their context"""
        negated_phrases = []

        for pattern in self.negation_patterns:
            matches = re.finditer(pattern, text, re.IGNORECASE)
            for match in matches:
                # Get the next few words after negation
                start = match.end()
                words_after = text[start:start+50].split()[:3]  # Next 3 words
                if words_after:
                    negated_phrases.append(' '.join(words_after))

        return negated_phrases

    def extract_enhanced_features(self, text):
        """Extract features including negation-aware keyword counts"""
        preprocessed = self.preprocess_text(text)
        negated_phrases = self.detect_negation_context(preprocessed)

        features = {}

        # Count keywords in each category
        for category, keywords in self.high_risk_keywords.items():
            count = 0
            negated_count = 0

            for keyword in keywords:
                # Count regular occurrences
                count += len(re.findall(r'\b' + keyword + r'\b', preprocessed))

                # Count negated occurrences
                for phrase in negated_phrases:
                    if keyword in phrase:
                        negated_count += 1

            features[f'high_risk_{category}'] = count
            features[f'high_risk_{category}_negated'] = negated_count
            features[f'high_risk_{category}_net'] = max(0, count - negated_count)

        # Same for moderate risk
        for category, keywords in self.moderate_risk_keywords.items():
            count = 0
            negated_count = 0

            for keyword in keywords:
                count += len(re.findall(r'\b' + keyword + r'\b', preprocessed))
                for phrase in negated_phrases:
                    if keyword in phrase:
                        negated_count += 1

            features[f'moderate_risk_{category}'] = count
            features[f'moderate_risk_{category}_negated'] = negated_count
            features[f'moderate_risk_{category}_net'] = max(0, count - negated_count)

        # Low risk keywords
        for category, keywords in self.low_risk_keywords.items():
            count = 0
            for keyword in keywords:
                count += len(re.findall(r'\b' + keyword + r'\b', preprocessed))
            features[f'low_risk_{category}'] = count

        # Additional features
        features['text_length'] = len(preprocessed)
        features['word_count'] = len(preprocessed.split())
        features['negation_count'] = len(negated_phrases)
        features['caps_ratio'] = sum(1 for c in text if c.isupper()) / len(text) if text else 0

        return features

    def create_rule_based_prediction(self, text):
        """Create rule-based prediction as baseline"""
        features = self.extract_enhanced_features(text)

        # Calculate risk scores
        high_risk_score = sum(v for k, v in features.items()
                             if k.startswith('high_risk_') and k.endswith('_net'))

        moderate_risk_score = sum(v for k, v in features.items()
                                 if k.startswith('moderate_risk_') and k.endswith('_net'))

        low_risk_score = sum(v for k, v in features.items()
                            if k.startswith('low_risk_'))

        # Enhanced decision rules
        if high_risk_score >= 3:
            return 'High Risk'
        elif high_risk_score >= 2:
            return 'High Risk'
        elif high_risk_score >= 1 and moderate_risk_score >= 1:
            return 'High Risk'
        elif high_risk_score >= 1 or moderate_risk_score >= 3:
            return 'Moderate Risk'
        elif moderate_risk_score >= 1:
            return 'Moderate Risk'
        elif low_risk_score >= 1:
            return 'Low Risk'
        else:
            return 'Moderate Risk'  # Default for unclear cases

    def analyze_negation_patterns(self, df):
        """Analyze negation patterns in the dataset"""
        negation_analysis = []

        for idx, text in enumerate(df['Comentarios']):
            preprocessed = self.preprocess_text(text)
            negated_phrases = self.detect_negation_context(preprocessed)

            if negated_phrases:
                negation_analysis.append({
                    'index': idx,
                    'incident_no': df.iloc[idx]['No. Incidente'],
                    'original_text': text,
                    'negated_phrases': negated_phrases,
                    'negation_count': len(negated_phrases)
                })

        return pd.DataFrame(negation_analysis)

# Load and analyze your data
def load_and_classify_data():
    """Load the real dataset and perform classification"""

    # Load your dataset
    print("Loading dataset: nuse_sample_50")
    df = nuse_sample_50.copy()  # Assuming the dataset is already loaded in Colab

    print(f"Dataset shape: {df.shape}")
    print(f"Columns: {df.columns.tolist()}")

    # Display basic info
    print("\n=== DATASET OVERVIEW ===")
    print(df.info())

    # Check for missing values in comments
    print(f"\nMissing values in 'Comentarios': {df['Comentarios'].isna().sum()}")

    # Initialize classifier
    classifier = DomesticViolenceClassifier()

    print("\n=== CLASSIFYING CASES ===")

    # Apply rule-based classification
    df['predicted_risk'] = df['Comentarios'].apply(classifier.create_rule_based_prediction)

    # Show distribution of predictions
    print("\nRisk Classification Distribution:")
    print(df['predicted_risk'].value_counts())

    # Analyze negation patterns
    print("\n=== NEGATION ANALYSIS ===")
    negation_df = classifier.analyze_negation_patterns(df)

    if not negation_df.empty:
        print(f"Cases with negation patterns: {len(negation_df)}")
        print("\nExamples of negated phrases:")
        for i, row in negation_df.head(5).iterrows():
            print(f"Case {row['incident_no']}: {row['negated_phrases']}")
    else:
        print("No negation patterns detected")

    return df, classifier, negation_df

# Detailed analysis functions
def analyze_high_risk_cases(df):
    """Analyze high-risk cases in detail"""
    high_risk_cases = df[df['predicted_risk'] == 'High Risk']

    print(f"\n=== HIGH RISK CASES ANALYSIS ({len(high_risk_cases)} cases) ===")

    for idx, case in high_risk_cases.iterrows():
        print(f"\nCase {case['No. Incidente']}:")
        print(f"Location: {case['Localidad']}")
        print(f"Date: {case['Fecha']} {case['Hora']}")
        print(f"Comment: {case['Comentarios']}")
        print(f"Risk Level: {case['predicted_risk']}")
        print("-" * 50)

def analyze_keyword_frequency(df, classifier):
    """Analyze keyword frequency across all cases"""

    print("\n=== KEYWORD FREQUENCY ANALYSIS ===")

    all_features = []
    for comment in df['Comentarios']:
        features = classifier.extract_enhanced_features(comment)
        all_features.append(features)

    feature_df = pd.DataFrame(all_features)

    # Sum up keyword categories
    high_risk_cols = [col for col in feature_df.columns if col.startswith('high_risk_') and col.endswith('_net')]
    moderate_risk_cols = [col for col in feature_df.columns if col.startswith('moderate_risk_') and col.endswith('_net')]

    print("Most frequent HIGH RISK indicators:")
    for col in high_risk_cols:
        total = feature_df[col].sum()
        if total > 0:
            print(f"  {col}: {total} occurrences")

    print("\nMost frequent MODERATE RISK indicators:")
    for col in moderate_risk_cols:
        total = feature_df[col].sum()
        if total > 0:
            print(f"  {col}: {total} occurrences")

def create_visualizations(df):
    """Create visualizations for the analysis"""

    plt.figure(figsize=(15, 10))

    # Risk distribution
    plt.subplot(2, 3, 1)
    risk_counts = df['predicted_risk'].value_counts()
    plt.pie(risk_counts.values, labels=risk_counts.index, autopct='%1.1f%%')
    plt.title('Risk Level Distribution')

    # Cases by locality
    plt.subplot(2, 3, 2)
    locality_counts = df['Localidad'].value_counts().head(10)
    plt.barh(locality_counts.index, locality_counts.values)
    plt.title('Cases by Locality (Top 10)')
    plt.xlabel('Number of Cases')

    # Risk by locality
    plt.subplot(2, 3, 3)
    risk_by_locality = df.groupby(['Localidad', 'predicted_risk']).size().unstack(fill_value=0)
    risk_by_locality.plot(kind='bar', stacked=True)
    plt.title('Risk Distribution by Locality')
    plt.xticks(rotation=45)
    plt.legend(title='Risk Level')

    # Text length distribution
    plt.subplot(2, 3, 4)
    df['comment_length'] = df['Comentarios'].str.len()
    plt.hist(df['comment_length'], bins=20, alpha=0.7)
    plt.title('Comment Length Distribution')
    plt.xlabel('Characters')
    plt.ylabel('Frequency')

    # Risk vs comment length
    plt.subplot(2, 3, 5)
    df.boxplot(column='comment_length', by='predicted_risk')
    plt.title('Comment Length by Risk Level')
    plt.xlabel('Risk Level')
    plt.ylabel('Comment Length')

    # Cases by hour
    plt.subplot(2, 3, 6)
    df['hour'] = pd.to_datetime(df['Hora'], format='%H:%M:%S').dt.hour
    hourly_counts = df.groupby('hour').size()
    plt.plot(hourly_counts.index, hourly_counts.values, marker='o')
    plt.title('Cases by Hour of Day')
    plt.xlabel('Hour')
    plt.ylabel('Number of Cases')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

def generate_detailed_report(df, classifier, negation_df):
    """Generate a comprehensive analysis report"""

    print("=" * 60)
    print("DOMESTIC VIOLENCE CASE CLASSIFICATION REPORT")
    print("=" * 60)

    print(f"Total cases analyzed: {len(df)}")
    print(f"Cases with negation patterns: {len(negation_df)}")

    # Risk distribution
    risk_dist = df['predicted_risk'].value_counts()
    print(f"\nRisk Distribution:")
    for risk_level, count in risk_dist.items():
        percentage = (count / len(df)) * 100
        print(f"  {risk_level}: {count} cases ({percentage:.1f}%)")

    # Geographic distribution
    print(f"\nGeographic Distribution:")
    locality_dist = df['Localidad'].value_counts().head(5)
    for locality, count in locality_dist.items():
        print(f"  {locality}: {count} cases")

    # High-risk cases summary
    high_risk_cases = df[df['predicted_risk'] == 'High Risk']
    if len(high_risk_cases) > 0:
        print(f"\n=== HIGH RISK CASES REQUIRING IMMEDIATE ATTENTION ===")
        for idx, case in high_risk_cases.iterrows():
            print(f"\nIncident: {case['No. Incidente']}")
            print(f"Location: {case['Localidad']}")
            print(f"Time: {case['Fecha']} {case['Hora']}")
            print(f"Details: {case['Comentarios'][:100]}...")

    # Recommendations
    print(f"\n=== RECOMMENDATIONS ===")
    high_risk_count = len(df[df['predicted_risk'] == 'High Risk'])
    moderate_risk_count = len(df[df['predicted_risk'] == 'Moderate Risk'])

    print(f"1. Immediate intervention required for {high_risk_count} high-risk cases")
    print(f"2. Follow-up needed for {moderate_risk_count} moderate-risk cases")
    print(f"3. Review negation patterns in {len(negation_df)} cases for accuracy")
    print(f"4. Consider additional resources for top localities: {', '.join(locality_dist.head(3).index)}")

# Main execution
if __name__ == "__main__":
    try:
        # Load and classify data
        df_classified, classifier, negation_analysis = load_and_classify_data()

        # Perform detailed analyses
        analyze_high_risk_cases(df_classified)
        analyze_keyword_frequency(df_classified, classifier)

        # Create visualizations
        create_visualizations(df_classified)

        # Generate comprehensive report
        generate_detailed_report(df_classified, classifier, negation_analysis)

        # Save results
        df_classified.to_csv('classified_domestic_violence_cases.csv', index=False)
        print("\nResults saved to 'classified_domestic_violence_cases.csv'")

    except NameError:
        print("Error: Dataset 'nuse_sample_50' not found.")
        print("Please ensure your dataset is loaded in Colab with this exact name.")
        print("You can load it using: nuse_sample_50 = pd.read_csv('your_file.csv')")
```
